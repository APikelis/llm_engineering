```python
import os
from dotenv import load_dotenv
from numpy.char import startswith
load_dotenv(override=True)

api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print('No Api key find, please check .env file')
if not api_key.startswith('sk-proj-'):
    print("API Key found but is incorrect format. Please check .env file")
else:
    print("API Key found and starts with sk-proj. All looks good ✅")
```

### 1. Endpoints

```python
import requests

headers = {'Authorization': f'Bearer {api_key}', "Content-Type": "application/json"}

payload = {
    "model": "gpt-5-nano",
    "messages": [{'role': 'user', 'content': 'Tell me a fun fact'}]
}

payload
```

```python
response = requests.post(
    "https://api.openai.com/v1/chat/completions",
    headers=headers,
    json=payload
)

response.json()
```

```python
#if throwing errors run the below
#print(response.status_code, response.text)
```

```python
response.json()["choices"][0]['message']['content'] #usual response line but json format
```

```python
from openai import OpenAI

openai = OpenAI() #client library so we dont have to fiddle with http endpoints

message = [
    {'role': 'user', 'content': 'Tell me a fun fact' }
]
response = openai.chat.completions.create(
    model = 'gpt-5-nano',
    messages = message
)

response.choices[0].message.content
```

```python
##google

google_api_key = os.getenv("GOOGLE_API_KEY")

GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"

gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=google_api_key)

response = gemini.chat.completions.create(
    model = "gemini-2.5-flash",
    messages=message
)

response.choices[0].message.content
```
